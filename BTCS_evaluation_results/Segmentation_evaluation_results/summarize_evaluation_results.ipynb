```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Evaluation for meta
re_meta_benign = json.load(open('/data/slurm/xuzt/ICA_data/meta_evaluation_results_benign.json'))
re_meta_malignant = json.load(open('/data/slurm/xuzt/ICA_data/meta_evaluation_results_malignant.json'))
```

```python
# re_meta_benign
```

```python
def get_average_acc(re):
    overall_acc = []
    for i in re:
        overall_acc.append(re[i]['acc'])
    average = np.mean(overall_acc)
    return average

def get_average_F1(re):
    overall_F1 = []
    count = 0
    for i in re:
        if not np.isnan(re[i]['F1']):
            overall_F1.append(re[i]['F1'])
        else:
            count += 1
    average = np.sum(overall_F1) / (len(re.keys()) - count)
    return average

def get_average_iou(re):
    overall_iou = []
    for i in re:
        overall_iou.append(re[i]['iou'])
    average = np.sum(overall_iou) / len(re) 
    return average

average_acc_meta_benign = get_average_acc(re_meta_benign)
average_F1_meta_benign = get_average_F1(re_meta_benign)
average_iou_meta_benign = get_average_iou(re_meta_benign)


average_acc_meta_malignant = get_average_acc(re_meta_malignant)
average_F1_meta_malignant = get_average_F1(re_meta_malignant)
average_iou_meta_malignant = get_average_iou(re_meta_malignant)

print('average_acc_meta_benign: ', average_acc_meta_benign)
print('average_F1_meta_benign: ', average_F1_meta_benign)
print('average_iou_meta_benign: ', average_iou_meta_benign)


print('average_acc_meta_malignant: ', average_acc_meta_malignant)
print('average_F1_meta_malignant: ', average_F1_meta_malignant)
print('average_iou_meta_malignant: ', average_iou_meta_malignant)
```
